# 1 Series基本介绍

类似一个一维数组，具有数据和索引

- 可以包含任意数据类型
- DataFrame的每一列都是Series
- 仅由索引和数据值组成
    - 索引：每一行的标签
    - 数据值，Series中的实际值


# 2 Series的创建

## 2.1 从列表创建

```python
import pandas as pd
data = [1,2,3,4]
s_data = pd.Series(data=data)
# 自定义索引
s_data = pd.Series(data=data,index=['a','b','c','d'])
s_data
```

## 2.2 从字典创建

键成为索引，值成为数据值

```python
data = {'a':[1,2,3,4]}
s_data = pd.Series(data=data)
s_data
```

```python
data = {'a':1,'b':2,'c':3,'d':4}
s_data = pd.Series(data=data)
s_data
```

# 3 主要方法

## 3.1 访问元素

### 3.1.1 按整数位置访问

类似列表，方括号加位置索引

### 3.1.2 按自定义索引

类似字典，方括号加自定义索引

```python
s_data[1:3]
```

```python
s_data['a']
```

### 3.1.3 通过.iloc[]方法访问

s.iloc[interger]:整数位置


### 3.1.4 通过.loc[]方法访问

s.iloc[index]:索引

```python
s_data.iloc[3]
```

```python
s_data.loc['b']
```

## 3.2 索引与值

Series由索引和值组成

### 3.2.1 index:获取索引和修改

```python
s_data.index
```

```python
s_data.index = ['k','l','m','n']
```

```python
s_data
```

### 3.2.2 values获取值

```python
s_data.values
```

# 4 DataFrame和Series关系
