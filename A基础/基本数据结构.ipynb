# 1 基本数据结构
包括列表list、元组tuple、集合set、字典dict
## 1.1 列表 list

- 可变，可修改
- 可包含不同类型元素
- 有序

```python
# 列表创建，方括号包裹元素
temp_list = [1,3.4,'34','6573','wys',[8,9,6]]
```

### 1.1.1 访问列表元素

#### 1.1.1.1 list[index]:访问单个元素，用索引，

从0开始，逆序则最后一个索引为-1，倒数第二个为-2

```python
temp_list[2]
```

```python
temp_list[-2]
```

#### 1.1.1.2 list[start,end,step]:访问多个元素，用切片

- start：起始索引，包含，默认为0
- end：结束索引，不包含，默认是结尾
- step：步长,默认为1
- 从第一个位置开始取，后面在原来索引值加step就是要取的元素
- 不写结尾或开头的索引默认取到后面的所有元素或从开头取
- 步长为负则逆序取值,此时索引必须按先后面再前面的顺序，不然为空

```python
temp_list = [1,3.4,'34','6573','wys',[8,9,6]]
temp_list[2:5:2]
```

```python
temp_list[2:]
```

```python
temp_list[:3]
```

```python
temp_list[4:1:-1]
```

```python
temp_list[1:4:-1]
```

```python
temp_list[::-2]
```

### 1.1.2 len(list)列表长度

```python
len(temp_list)
```

### 1.1.3 in/not in 元素是否在列表中

```python
temp_list = [1,3.4,'34','6573','wys',[8,9,6]]
1 in temp_list
```

```python
33 in temp_list
```

```python
33 not in temp_list
```

### 1.1.4 append(): 列表中末尾添加**一个**元素

> extend()一次性添加多个值



```python
temp_list.append('hsj')
temp_list.append(['w','d','90'])
temp_list
```

### 1.1.5 extend([]) 在列表末尾一次性添加多个值

> append() 只添加一个元素

```python
temp_list.extend(['shsjahs','sss','445'])
temp_list
```

### 1.1.6 insert(index, value) 在列表指定位置添加***一个***值

索引从0开始，比如位置1，添加到位置1，原来元素依次后移

```python
my_list = [1,2,3]
my_list.insert(1,"a")
my_list
```

```python
my_list.insert(1,[4,5,6])
```

### 1.1.7 remove(value) 移除列表中第一个值为value的元素

```python
my_list.remove('a')
my_list
```

### 1.1.8 pop(5) 移除列表中指定的元素并返回,默认是最后一个

```python
my_list = [1,2,3,4,'a',6]
value = my_list.pop(1)
print(value)
```

### 1.1.9 index(value) 返回列表第一个value的索引

```python
my_list = [1,2,3,4,'a',6]
my_list.index(3)
```

### 1.1.10 count(value) 统计列表出现value的次数


```python
my_list.count(1)
```

### 1.1.11 sort(reverse=False) 列表元素排序

reverse = True 降序，False表示升序，元素之间必须是可比较的

> sorted()函数，通用的排序方法，可指定关键字

```python
my_list = [1,9,3,4,7,6]
my_list.sort(reverse=False)
my_list
```

### 1.1.12 reverse() 反转列表


```python
my_list = [1,9,3,4,7,6]
my_list.reverse()
my_list
```

## 1.2 元组 tuple

- 不可变
- 可包含不同类型元素
- 有序的

```python
# 元组创建，小括号包围
(1,2,3)
```

### 1.2.1 访问元组元素
#### 1.2.1.1 tuple[index]访问单个元素

```python
my_tuple = (1,2,3,4)
my_tuple[1]
```

#### tuple[start,end,step]访问多个元素

- start:起始，包括该位置，默认为0
- end:结尾，不包括，默认到元素末尾
- step：步长，默认为1.步长为负数可以反向访问

> 类似列表

```python
my_tuple = (1,2,3,4,5)
my_tuple[1:4:2]
```

```python
my_tuple[3::]
```

```python
my_tuple[::-1]
```

### 1.2.2 count(x) 统计某个元素在元组中出现次数

```python
my_tuple.count(1)
```

### 1.2.3 index(x) 返回第一个x的索引，索引从0开始

```python
my_tuple = (1,2,3,4,5)
my_tuple.index((3))
```

## 1.3 字典dict

特性
- 可变
- 无序的
- 键值对存储数据，key是不可变的

```python
# 创建字典，花括号包围键值对，逗号分割
my_dict = {'a':1,'b':2,'c':3,'d':4,'e':5}
```

### 1.3.1 访问字典元素，通过键

```python
my_dict = {'a':1,'b':2,'c':3,'d':4,'e':5}
my_dict['a']
```

### 1.3.2 get(key, default=None):返回指定键的值，如果键不在则返回默认值

```python
my_dict.get('a')
```

```python
my_dict.get('x', "No result")
```

### 1.3.3 keys()：返回一个字典的键的列表

```python
my_dict.keys()
```

### 1.3.4 values():返回一个字典的值的列表

```python
my_dict.values()
```

### 1.3.5 items(): 返回一个字典键值对的列表

[(key1,v1),(k2,v2),(k3,v3)]

```python
my_dict.items()
```

### 1.3.6 update({key:value}): 将键值对添加到列表中

```python
my_dict.update({'f':6,'g':7})
my_dict
```

### 1.3.7 pop(key,default):移除对应的键值对，并返回其值

```python
my_dict.pop('d')
```

```python
my_dict
```

### 1.3.7 popitem(key,default):移除最新添加的键值对，并返回其键-值对的元组

```python
my_dict.popitem()
```

## 1.4 集合set

特性
- 可变的
- 无序的
- 元素不重复的
- 无法直接使用索引访问

```python
# 创建，花括号包围
{1,2,3}
```

### 1.4.1 访问集合元素

#### 访问单个元素，转换为列表或使用for循环

```python
my_set = {'a', 'b', 'c', 'd'}
for i in my_set:
    if i=='a':
        print(i)
```

### 1.4.2 add(x)：添加元素

```python
my_set.add('x')
my_set
```

### 1.4.3 remove(x):删除元素

```python
my_set.remove('x')
my_set
```

### 1.4.4 clear():清空集合

```python
my_set.clear()
```

### 1.4.5 | 并集; & 交集； - 差集; ^ 对称差集

```python
set1 = {1,2,3}
set2 = {3,5,6}
set1 | set2
```

```python
set1 & set2
```

```python
set1 - set2
```

```python
# 只在两个集合中分别存在但不同时出现的
set1 ^ set2
```

### 1.4.6 issubset()：判断是否子集

```python
set1.issubset(set2)
```

```python
set1.issubset({1,2,3,4})
```

### 1.4.7 issuperset():判断是否超集

```python
set1.issuperset({1,2})
```

```python
set1.issuperset({1,2,5})
```

### 1.4.8 isdisjoint()：判断两个集合是否相交

```python
set1.isdisjoint(set2)
```

### 1.4.9 len()：计算集合长度

```python
len(set1)
```

### 1.4.10 in/not in ：判断元素是否在集合中

```python
2 in set1
```

```python
6 in set1
```

# 2 基本数据结构转换

## 2.1 列表和元组

- list()：转列表
- tuple():转元组

```python
set1 = {1,2,3}
list(set1)
```

```python
set([1,2,3])
```

## 2.2 列表和字典

- dict（）：转字典

```python
my_list = [(1,'a'),(2,'b'),(3,'c')]
dict1 = dict(my_list)
```

```python
# 字典的键或值或键值对转列表
list(dict1.keys())
list(dict1.values())
list(dict1.items())
```

## 2.3 列表转集合

set():转换为集合，会对元素去重

```python
my_list = [(1,'a'),(2,'b'),(3,'c')]
set(my_list)
```

## 2.4 字典和集合

dict（）：转换为字典
dict.fromkeys():将集合转换为字典的键，值默认为None

```python
my_set = {(1, 'a'), (2, 'b'), (3, 'c')}
my_dict = dict(my_set)
```

```python
# # 字典的键或值或键值对转集合
set(my_dict.items())
set(my_dict.keys())
set(my_dict.values())
```

```python
my_dict.fromkeys({1,2,3})
```

## 2.5 字典和元组

tuple():转换为元组

```python
my_dict
```

```python
tuple(my_dict.items())
tuple(my_dict.keys())
tuple(my_dict.values())
```

```python
tuple(my_dict)
```

```python
my_tuple = ((1, 'a'), (2, 'b'), (3, 'c'))
dict(my_tuple)
```

## 2.6 元组转集合

set（）：会对元素去重

tuple（）

```python
my_tuple = ((1, 'a'), (2, 'b'), (3, 'c'))
set(my_tuple)
```

```python
my_set = {(1, 'a'), (2, 'b'), (3, 'c')}
tuple(my_set)
```
