# 1 函数

## 1.1 特点和设计原则

- 封装性：操作封装，减少冗余
- 复用性，避免重复
- 模块化，程序逻辑分为独立模块，结构清晰
- 可维护性

## 1.2 函数的基本结构

```python
# 定义
def func_name(parameters):

    process
    return something

# 调用函数，传入参数，返回值赋值
x = func_name(parameters='xxx')

```

````python
def my_func():
    print("hello world")
my_func()
````

## 1.3 函数的分类

- 无参函数：不用传参，自动执行
- 有参函数：
- 有返回值
- 无返回值

## 1.4 参数类型

### 1.4.1 位置参数，按照顺序传递参数

不写参数名，直接写参数值

````python
def subract(a,b):
    return a-b
subract(1,2)
````

### 1.4.2 关键字参数，可以不按顺序，通过参数名传递

> 关键字参数必须放在位置参数之后

````python
def mood(name,mod):
    return f"I'm {name} and I'm {mod}"
mood(name='x',mod=2)
````

### 1.4.3 默认参数，为参数设置默认值，可以不传参数

````python
def mood(name='df',mod='ddd'):
    return f"I'm {name} and I'm {mod}"
mood(name='x',mod='er')
````

````python
mood()
````

### 1.4.4 可变参数，参数数量不确定

#### 1.4.4.1 *args：可变位置参数，将参数收集为一个元组

> 函数体里直接用args，不用加*


````python
def mood(*args):
    print(args)
mood(1,2,3,4)
````

#### 1.4.4.1 **kwargs：可变关键字参数，将参数收集为一个字典

> 函数体里直接用kwargs，不用加**

````python
def mood(**kwargs):
    print(kwargs)
mood(x=1,y=2,z=3,t=4)
````

### 1.5 函数作用域


#### 1.5.1 局部变量

只能在函数内部使用

#### 1.5.2 全局变量

整个程序，函数内外都可用

#### 1.5.3 global关键字

- 在函数内部定义的变量在外部也可用
- 先声明，后赋值

````python
def mood(name='df',mod='ddd'):
    global x
    x = 10
    return f"I'm {name} and I'm {mod}"
mood(name='x',mod='er')
print(x)
````

# 2 匿名函数

不用写函数名称，简单

## 2.1 匿名函数语法

lambda 参数1，参数2：表达式

> 表达式不能包含多行，匿名函数返回表达式的值

优点

- 简洁，不用对太简单的函数写太多内容
- 灵活，可直接嵌入其他函数，如sorted、map、apply
- 可读性

````python
(lambda x,y,z: x+y+z)(2,3,4)
````

````python
lambda_func = lambda x,y,z: x+y+z
lambda_func(1,2,3)
````

## 2.2 匿名函数使用场景

### 2.2.1 直接使用，见上，有另外赋值和不赋值两种

### 2.2.2 与sorted函数结合

````python
# 按第二个值排序
data = [('a',2),('b',0),('c',5),('d',4)]
sorted_data = sorted(data,key=lambda x:x[1],reverse=True)
sorted_data
````

### 2.2.2 与map函数结合

````python
# map函数对列表中的每个元素应用一个函数
nums = [1,2,3,4,5]
list(map(lambda x:x**2, nums))
````

### 2.2.3 与列表表达式结合

````python
# 列表表达式，快速生成新列表
nums = [1,2,3,4,5]
squares = [lambda x:x**2 for x in nums]
squares
````

````python
squares = [(lambda x:x**2)(x) for x in nums]
squares
````

# 3 异常

语法
else是正常执行会触发的，finally是不论如何最后都会执行的

````python
def test(num):
    try:
        out = 10/num
    except Exception as e:
        # 捕获所有错误
        print(e)
    else:
        print(f"结果：{out}")
        return out
    finally:
        print("执行结束")

test(0)
````

````python
test(1)
````
