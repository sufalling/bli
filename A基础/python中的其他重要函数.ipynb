# 1 print()

## 1.1 基本语法

print(objects,sep='',end='\n')

- object:打印对象
- sep:多个内容使用的分隔符，默认空格
- end:输出结果末尾字符，默认换行\n

### 1.1.1 输出多个值


```python
# 以-----分隔多个值
print('a','b',sep='-----',end='')
print('c','a',sep='-----',end='')
```

## 1.2 格式化print()输出

将变量插入字符串实现动态输出

### 1.2.1 format

'str{}'.format(values)

- {}:变量名
- values：变量值

#### 1.2.1.1 直接传值（顺序传入）

```python
name = 'alice'
age = [20,21,11]
for i in age:
    print("my name is {} and my age is {}".format(name,i))
```

#### 1.2.1.2 使用位置索引

花括号占位符，里面是format函数中参数的位置

```python
name = 'alice'
age = [20,21,11]
print("my name is {1} and my age is {0}".format(name,age))
```

#### 1.2.1.3 使用关键字参数

花括号里写键，format括号里为{}中的键赋值

```python
name = 'alice'
age = 20
print("my name is {name} and my age is {age}".format(name=name,age=age))
```

#### 1.2.1.4 数字格式化

- {:.2f]:两位小数
- {:,}数字分位符
- {:+}正负号,实际正负看数字的正负
- {:e}科学计数法



```python
num = 12345.1234
num2 = -1234.222
print('this is num of {:.2f}'.format(num))
```

```python
print('this is num of {:e}'.format(num))
```

```python
print('this is num of {:+}'.format(num2))
```

```python
print('this is num of {:,}'.format(num))
```

### 1.2.2 f-string

f"str{var}",使用f或F开头，Python会自动转换合适的数据类型,可以用上述格式化的各种条件

{}可以填变量或表达式或函数

```python
import math
age = 21
f"i'm {age:.2f}，next year i will be {age + 1},pi is {math.pi}"
```

# 2 random

用于生成随机数或随机选择


| 函数                                | 描述                                 |
|-----------------------------------|------------------------------------|
| random.seed()                     | 随机数种子                              |
| random.random()                   | 返回【0，1）范围内随机浮点数                    |
| random.uniform(a,b)               | 返回【a,b】范围内随机浮点数，包括a和b              |
| random.randint(a,b)               | 返回【a,b】范围内随机整数，包括a和b               |
| random.randrange(start,stop,step) | 返回【a,b）范围内步长为step的整数，包括a不包括b       |
| random.choice(sequence)           | 从非空序列如列表元组字符串等随机选择一个元素             |
| random.choices(seqence,k=N)       | 从非空序列如列表元组字符串等随机选择N个元素，允许重复，返回一个列表 |
| random.sample(seqence,k)          | 从非空序列如列表元组随机选择N个元素，不允许重复，返回一个列表    |
| random.shuffle(seqence)           | 打乱可变序列的元素（字符串不行），直接在原序列上修改         |

## 2.1 random.seed():随机数种子，确保结果可复现

```python
import random
random.seed(1)
```

## 2.2 random.random()生成0-1随机数

```python
random.random()
```

## 2.3 random.uniform(a,b)生成[a,b]范围内随机浮点数

```python
random.uniform(1,2)
```

## 2.4 random.randint():生成[a,b]范围内随机整数

```python
random.randint(1,10)
```

## 2.5 random.randrange()：生成[a,b)范围内步长为step的随机整数

```python
# 生成被3整除的随机数
random.randrange(0,50,3)
```

## 2.6 random.choice():从序列中随机选择一个元素



```python
random.choice([1,9,3,8,0])
```

```python
random.choice('jspoll')
```

## 2.7 random.choices():从序列中随机选择k个元素，可重复

```python
random.choices('qwertyuio',k=4)
```

## 2.8 random.sample():从序列中随机选择k个元素，不许重复

```python
random.sample('qwertyui',k=5)
```

## 2.9 random.shuffle():随机打乱序列，函数本身不返回任何内容

```python
l = [1,2,3,4,5,6]
random.shuffle(l)
l
```
