# 1 ls.sort(key,reverse):对列表进行排序，改变原始数据

- keys: 从每个元素提取排序的键,传入函数名，
    - 函数的输入看作是每个元素,可以传入元组，代表按从左到右的多个值进行排序,前一个条件一致，再根据后一个条件排序
- reverse:True，降序；False升序，默认升序

```python
nums = [1,2,3,4,5,6,7]
# reverse
nums.sort(reverse=True)
nums
```

```python
# 按字符串长度排序
fruit = ['appleeeee','banana','orange']
fruit.sort(key=len)
fruit
```

```python
people_l = [
    {'name':'alice','age':80},
    {'name':'bob','age':12},
    {'name':'charil','age':23},
    {'name':'dad','age':34},
]

people_l.sort(key=lambda x:x['age'])
people_l
```

```python
people_l = [
    {'name':'alice','age':80,'score':250},
    {'name':'alice','age':80,'score':251},
    {'name':'bob','age':12,'score':209},
    {'name':'charil','age':23,'score':201},
    {'name':'dad','age':34,'score':214},
    {'name':'dad','age':33,'score':214},
]

people_l.sort(key=lambda x:(x['age'],x['score']))
people_l
```

# 2 sorted(iterable,key,reverse)

对任何可迭代对象进行排序，返回一个新排序的列表，不改变原始数据

- iterable:元组、列表、字符串
- key:提取排序键的函数
- reverse:False，默认，升序；True，降序

```python
nums = [1,2,3,4,5,6,7]
sorted(nums,reverse=True)
```

```python
people_l = (
    {'name':'alice','age':80,'score':250},
    {'name':'alice','age':80,'score':251},
    {'name':'bob','age':12,'score':209},
    {'name':'charil','age':23,'score':201},
    {'name':'dad','age':34,'score':214},
    {'name':'dad','age':33,'score':214},
)

sorted(people_l,key=lambda x:(x['age'],x['score']),reverse=True)
```
