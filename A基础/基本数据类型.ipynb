# 1 基本数据类型

## 1.1 基本数据类型介绍

包括整数（int）、浮点（float）、布尔（Boolean）、字符串（str）

### 1.1.1 整数（int）

常用方法

#### 1.1.1.1 abs():返回绝对值

```python
abs(-10)
```

#### 1.1.1.2 pow(base, exp):返回幂
 $ {base} ^ {exp} $


```python
pow(2,2)
```


### 1.1.2 浮点（float）

#### 1.1.2.1 abs():绝对值

```python
abs(-34.222)
```

#### 1.1.2.2 round():四舍五入

```python
round(23.3341)
```

#### 1.1.2.3 math.floor():向下取整

```python
import math
math.floor(23.3341)
```

#### 1.1.2.4 math.cell():向上取整

```python
math.ceil(23.3341)
```


### 1.1.3 字符串（str）

**特性：**
- 表示字符序列，引号包裹
- 可以包含字母、数字、符号等

#### 1.1.3.1 capitalize():首字母大写

```python
"hello world".capitalize()
```

#### 1.1.3.2 upper():全部字母大写

```python
"hello world".upper()
```

#### 1.1.3.3 lower():全部字母小写


```python
"HELLO WORLD".lower()
```

#### 1.1.3.4 find(substring):查找子字符串的位置

索引从0开始,返回子字符串开始匹配的位置，空格算一个字符

```python
"hello， world".find("wo")
```

#### 1.1.3.5 replace(old,new):替换成新字符

```python
"world".replace("wo","ho")
```

#### 1.1.3.6 split(delimiter):安照分隔符分割字符，返回列表

```python
"hello， world".split(', ')
```

#### 1.1.3.7 join(iterable)：将可迭代对象中的元素按照某个字符连接成一个字符串

```python
','.join(['hello','world','python'])
```

#### 1.1.3.8 strip():移除字符串头尾的字符，无参数默认为空格

如果有多个字符，则分别去除涉及的每个字符

```python
'hello, world'.strip('d')
'hello, world'.strip('od')
```

```python
'hello, world'.strip('hd')
```

#### 1.1.3.9 字符串格式化


```python
name1 = 'sss'
name2 = "dfd"
f"shshsh{name1} and {name2}"
"shshsh{n1} and {n2}".format(n1=name1,n2=name2)
"shshsh{0} and {1}".format(name2,name1)
```

#### 1.1.3.10 str.endwith('a'):寻找以a结尾的字符串，返回布尔值

```python
'sshajjs'.endswith('s')
```


### 1.1.4 布尔（Boolean）

- 只有1，0两个取值
- 逻辑运算 and or not
- 只要元素不为空或0，都返回True

```python
bool(12)
```


## 1.2 基本数据类型之间转换

- str(x):转换为字符串
- int(x):转换为整数
- float(x):转换为浮点数
- bool(x):转换为布尔值

### 1.2.1 整数转浮点数

```python
int_num = 2
float(int_num)
```

### 1.2.2 浮点数转整数

```python
float_num = 2.3
int(float_num)
```

### 1.2.3 字符串转整数,整数转字符串

```python
int('23')
```

```python
str(23)
```

### 1.2.4 整数转布尔值，0为False，其他为True

```python
print(bool(12),bool(0))
```

### 1.2.5 布尔值转整数


```python
print(int(True),int(False))
```

### 1.2.6 浮点数转字符串


```python
str(23.11)
```

### 1.2.7 字符串转浮点数


```python
float("23.54")
```

### 1.2.8 浮点数转布尔值，0为False，其他为True


```python
bool(0)
```

```python
bool(23.11)
```

### 1.2.9 布尔值转浮点数，False=0，True=1.0

```python
float(True)
```

```python
float(False)
```

### 1.2.10 字符串转布尔值，空为False，其他为True

```python
bool("shsj")
```

```python
bool("")
```
