# 1 路径概念

文件存放位置

## 1.1 相对路径
相对于当前工作目录的路径，从当前根目录出发，描述文件或目录的位置

常用标识符
- . 当前目录
- .. 父级目录

例如当前目录home/user ,
- ./document/test.md表示home/user/document/test.md
- 。。/document/test.md表示home/document/test.md

## 1.2 绝对路径

从系统根目录出发，类unix系统是/，Windows是驱动器盘符，无论在哪里绝对路径指向唯一的位置

# 2 os模块的路径操作

## 2.1 os.getcwd()：获取当前目录


```python
import os
```

```python
os.getcwd()
```

## 2.2 os.listdir(path)：展示路径下的所有文件和文件夹名称

```python
os.listdir('./')
```

## 2.3 os.path.join(path1,path2,...):拼接路径



```python
# 遍历文件
import os

dir = '../test'
for file_name in os.listdir():
    print(os.path.join(dir,file_name))
```

## 2.4 os.path.abspath(path):获取路径的绝对路径

注意当前的工作路径，该函数不会检查绝对路径是否正确

```python
os.path.abspath('../test')
```

## 2.5 os.mkdir(path):创建文件夹

```python
print(os.getcwd())
os.mkdir('./tmp')
```

## 2.6 os.rmdir(path):移除文件夹

```python
os.rmdir('./tmp')
```

## 2.7 os.path.exists(path):判断路径是否存在

```python
os.path.exists('./tmp')
```

```python
os.path.exists('../test/test.md')
```

```python
# 使用，常和文件夹创建合用

dir_path = '../test/test2'
if not os.path.exists(dir_path):
    os.mkdir(dir_path)
    print('创建文件夹')
else:
    print('已经存在')
```

## 2.8 os.path.isdir(path):判断是否是文件夹

不存在，不是文件夹--False

```python
os.path.isdir('./test')
```

```python
os.path.isdir('../test/test2')
```

## 2.8 os.path.isfile(path):判断是否是文件夹

不存在，不是文件--False

```python
os.path.isfile('./基本语句.ipynb')
```

```python
os.path.isfile('../test')
```

 如果判断是文件，直接读取，如果判断是文件夹，使用os.listdir()和os.path.join()获得文件路径再读取
