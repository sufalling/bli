# 1 python运算符号

## 1.1 算术运算符


| 运算符 | 符号 | 示例   | 结果 |
|-----|----|------|:--:|
| 加   | +  | 1+2  | 3  |
| 减   | -  | 2-1  | 1  |
| 乘   | *  | 1*2  | 2  |
| 除   | /  | 2/1  | 2  |
| 整除  | // | 3//4 | 0  |
| 取余  | %  | 5%4  | 1  |
| 指数  | ** | 3**2 | 9  |


## 1.2 比较运算符，返回布尔值

| 运算符  | 符号 | 示例   |  结果   |
|------|----|------|:-----:|
| 等于   | == | 1==2 | False |
| 不等于  | != | 2!=1 | True  |
| 大于   | >  | 1>2  | False |
| 大于等于 | >= | 2>=1 | True  |
| 小于   | <  | 3<4  | True  |
| 小于等于 | <= | 5<=4 | False |


## 1.3 逻辑运算符，布尔值之间的运算

| 运算符  | 符号  | 示例             |  结果   |
|------|-----|----------------|:-----:|
| 或    | or  | True or False  | True  |
| 与    | and | True and False | False |
| 非    | not | not True       | False |

> | & - ^ 是集合的并，交，差，对称差运算，也有些是位运算符号


## 1.4 赋值运算符

| 运算符 | 符号  | 示例    |   结果   |
|-----|-----|-------|:------:|
| 赋值  | =   | x=2   |  x=2   |
| 加赋值 | +=  | x+=2  | x=x+2  |
| 减赋值 | -=  | x-=1  | x=x-1  |
| 乘赋值 | *=  | x*=2  | x=x*2  |
| 除赋值 | /=  | x/=1  | x=x/1  |
| 整除赋值 | //= | x//=4 | x=x//4 |
| 取余赋值 | %=  | x%=4  | x=x%4  |
| 幂赋值 | **= | x**=2 | x=x**2 |

## 1.5 成员运算符

判断序列中是否存在某个值

| 运算符   | 符号     | 示例              |  结果   |
|-------|--------|-----------------|:-----:|
| 在序列中  | in     | 1 in [2，3]      | False |
| 不在序列中 | not in | 2 not in [3, 4] | True  |


## 1.6 身份运算符

判断两个对象内存地址是否一致

| 运算符    | 符号     | 示例         |     结果     |
|--------|--------|------------|:----------:|
| 是同一对象  | is     | A is B     | True/False |
| 不是同一对象 | is not | A is not B | True/False |

## 1.8 运算优先级

| 运算符        | 优先级 |

| 指数         | 1   |

| ^          | 6   |
