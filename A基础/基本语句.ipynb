## 1.1 条件语句

- condition必须是布尔表达式，结果是布尔值
- 正确缩进，tab
- 从上到下，一旦满足条件，后续不再执行
- 可以嵌套

```python
if condition1:
    ...
elif condition2:
    ...
else:
    ...
```

## 1.2 循环语句

### 1.2.1 for循环

迭代次数固定

````python
for i in range(10):
    print(i)
````


### 1.2.2 while循环

- 条件为真一直执行，直到条件不成立
- 适合不确定次数的情况
- 注意设置中断条件，避免死循环

````python
i = 0
while i < 5:
    print(i)
    i += 1
````

### 1.2.3 控制循环

- break:立即终止循环
- continue：跳过该次循环
- pass：占位，空代码，保证结构完整。没有实际作用，不报错，方便后续添加

### 1.2.4 range(start,end,step):与for循环结合使用

- start:开始，包括，默认为0
- end:结尾，不包括
- step:步长，默认为1,如果为负数则每次递减

````python
for i in range(1,5,2):
    print(i)
````

````python
for i in range(5,3,-1):
    print(i)
````

### 1.2.5 enumerate

为可迭代对象添加索引，enumerate（iterable，start=0）

- iterable：可以是列表，元组，字符串
- start：索引的起始值，默认为0


````python
# 遍历一个列表
my_list = ['apple','xiaomi','huawei','oppo']
for index,value in enumerate(my_list):
    print(index,value)
````

````python
# 遍历多个列表，结合zip
corporate = ['apple', 'xiaomi', 'huawei', 'oppo']
boss = ['cock','Lei','Ren','Liu','no']
for index, item in enumerate(zip(corporate,boss),start=1):
    print(index,item)
````

## 1.3 tqdm:进度条显示

tqdm(iterable, desc)

- iterable:迭代对象
- desc：描述文字

````python
from tqdm.notebook import tqdm
cnt = 1
for i in tqdm(range(10**7),desc='test'):
    cnt += 1
print(cnt)
````
