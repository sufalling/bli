# 1 t分布

## 1.1 基本概念

t分布是一种对称、钟形的概率分布，主要用于以下场景

- 样本量较小（n<30,大于30用正态分布代替)
- 总体标准差未知

t分布在小样本的统计推断中非常重要，如：
- 单样本均值检验（t检验）
- 均值差异检验（成对样本、双样本）
- 构造置信区间（当标准差未知）

分布记法：
$$
t \sim t(v)
$$

-  其中v是自由度，v = 样本数 - 1

## 1.2 概率密度函数

表达式

$$
f(t) = \frac{\Gamma(\frac{v + 1}{2})}{\sqrt{v\pi}\Gamma(\frac{v}{2})}(1 + \frac{t^2}{v})^{-\frac{v+1}{2}}
$$

其中
$\Gamma$是伽马函数，v是自由度

![img](../picture/E-t-1.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t,norm
import seaborn as sns
sns.set_style('darkgrid')
# 自由度设置
dfs = [1,2,5,10,30,100]

# 创建x值范围
# 指定范围内生成等间隔数列,范围+样本量
x = np.linspace(-5,5,500)

# 计算标准正态分布的PDF
pdf = norm.pdf(x,0,1)

# 绘图
plt.figure(figsize=(10,6))
for df in dfs:
    plt.plot(x,t.pdf(x,df),label='t {}'.format(df))

# 添加参考线（标准正态分布）
plt.plot(x,norm.pdf(x,0,1),'k--',label='Standard Normal')

plt.legend()
plt.grid(True)
plt.show()
```
