# 1 正态分布

又称高斯分布，是最重要的连续概率分布之一，描述大量自然现象（身高、测量误差、考试成绩）的分布规律，特点为：

- 对称性：关于均值对称
- 集中性：数据集中在均值附近，远离均值$\mu$的概率逐渐降低
- 完全由参数决定，仅需$\mu$和$\sigma$即可确定分布形态，记作

$$
X \sim N(\mu, \sigma)
$$

## 1.2 概率密度函数

正态分布的概率密度函数（PDF）为

$$
f(x) =  \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

- \mu:均值，决定分布的中心位置
- \sigma:标准差，决定曲线胖瘦
- $e$:自然常数（$\approx 2.71826$）

![img](../picture/E-N-1.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

# 设置不同参数组合
params = [
    (0,1),
    (0,2),
    (2,1),
    (2,0.5),
]

# 创建x值范围
# 指定范围内生成等间隔数列,范围+样本量
x = np.linspace(-6,6,1000)

# 绘图
plt.figure(figsize=(10,6))

# 绘制多个正态分布曲线
for mu, sigma in params:
    y = norm.pdf(x,mu,sigma)
    plt.plot(x,y,label=f"μ={mu}, σ={sigma}")

# 添加参考线（标准正态分布）
plt.plot(x,norm.pdf(x,0,1),'k--',label='Standard Normal')
plt.legend()
plt.grid(True)
plt.show()
```

# 2标准正态分布
![img](../picture/E-N-2.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

# 设置不同参数组合
params = [
    (0,1),
    (0,2),
    (2,1),
    (2,0.5),
]

# 创建x值范围
# 指定范围内生成等间隔数列,范围+样本量
x = np.linspace(-4,4,500)

# 计算标准正态分布的PDF
pdf = norm.pdf(x,0,1)

# 绘图
plt.figure(figsize=(10,6))
plt.plot(x,pdf,label='Standard Normal')

# 添加3σ原则可视化
plt.fill_between(x,pdf,where=(x >= -1) & (x <= 1), color='skyblue', alpha=0.4, label='μ+1σ')
plt.fill_between(x,pdf,where=(x >= -2) & (x <= 2), color='orange', alpha=0.2, label='μ+2σ')
plt.fill_between(x,pdf,where=(x >= -3) & (x <= 3), color='green', alpha=0.1, label='μ+3σ')
plt.legend()
plt.grid(True)
plt.show()
```
