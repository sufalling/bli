## 1.4 性质

- 二元性：只能取0或1
- 极值特性：当p=0.5,方差最大（0.25）
- 可叠加性：n次独立伯努利实验的和服从二项分布B（n，p）

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import seaborn as sns
sns.set_style('darkgrid')
# 设置成功概率
probs = [0,2]
# 样本量
sample_size = 100



# 绘图
plt.figure(figsize=(10,6))

for p in probs:

    # 生成伯努利随机变量
    data = bernoulli.rvs(p=p, size=sample_size)
    # 使用countplot展示频数分布
    sns.countplot(x=data,label='Bernoulli',
                  alpha=0.7,
                  edgecolor='black')

plt.legend()
plt.grid(True)
plt.show()
```
