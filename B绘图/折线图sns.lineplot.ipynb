```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = sns.load_dataset('flights')
data
```

```python
# 加载数据集

# 设置Seaborn风格和颜色调色板
sns.set_style('whitegrid')
# 设置图片大小,单位英寸
plt.figure(figsize=(10,10))

# 绘制折线图
sns.lineplot(
    data=data,
    x='year',
    y='passengers',
    color='green',
)
# 图标标题，坐标名
plt.title('year vs passengers')
plt.xlabel('year')
plt.ylabel('passengers')
plt.show()
```

## 1.3 errorbar控制


参数=

None：不显示

ci：显示置信区间，（‘ci',95)表示95%置信区间

sd：显示标准差

se：显示标准误

标准差：

反映数据点与均值的偏差，反映波动程度

$$
\sigma = \sqrt{\frac{1}{N}\sum(x_i - \mu) ^ 2}
$$

标准误：

是样本均值的标准差，衡量样本均值的不确定性或估计误差，反映了如果多次抽样样本均值与真实总体均值直接的偏差，公式如下

$$
SE = \frac{\sigma}{\sqrt{n}}
$$

$ \sigma $是样本标准差，n是样本大小

```python
sns.set_style('white')

sns.lineplot(
    data=data,
    x='year',
    y='passengers',
    color='green',
    marker='o',
    linestyle='-',
    linewidth=4,
    errorbar=('ci',95)# 95%置信区间
)
plt.show()
```

```python
sns.set_style('white')

sns.lineplot(
    data=data,
    x='year',
    y='passengers',
    color='green',
    marker='o',
    linestyle='-',
    linewidth=4,
    errorbar='sd'# 95%置信区间
)
plt.show()
```

```python
sns.set_style('white')

sns.lineplot(
    data=data,
    x='year',
    y='passengers',
    color='green',
    marker='o',
    linestyle='-',
    linewidth=4,
    errorbar='se'# 95%置信区间
)
plt.show()
```

## 1.4 多条折线图

在渲染前设置多个sns.lineplot，并且添加label图例标签

```python

data = sns.load_dataset('flights')

sns.set_style('darkgrid')

# 选择感兴趣的数据集
data_1949 = data[data['year'] == 1949]
data_1955 = data[data['year'] == 1955]
data_1957 = data[data['year'] == 1957]

# 1949年折线图
sns.lineplot(
    data=data_1949,
    x='month',
    y='passengers',
    marker='o',
    linestyle='-',
    linewidth=2,
    label='1949',
)
# 1955
sns.lineplot(
    data=data_1955,
    x='month',
    y='passengers',
    marker='^',
    linestyle='--',
    linewidth=2,
    label='1955',
)

# 1968
sns.lineplot(
    data=data_1957,
    x='month',
    y='passengers',
    label='1957',
    marker='x',
    linestyle='-.',
    linewidth=2,
)

plt.xlabel('month')
plt.ylabel('passengers')

# 显示图例
# 需要在sns.line里设置label参数
plt.legend(title='Year')

plt.show()
```
