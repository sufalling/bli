# 1 山脊图

Ridgeline plot将每个类别的分布图（如直方图或he密度估计图）堆叠在一起，形成山脊一样的图，对多个变量进行可视化展示

## 1.1 适用场景

- 多分布比较：一个图表展示多个数据类别的分布
- 数据探索：曲线，探索数据分布趋势、差异、异常值、数据重叠、分布集中


## 1.2 joypy.joyplot


- data:
- column:需要绘制的变量名称
- by:按照某个字段对数据进行分组，可以按组绘制不同的子图
- fade:控制曲线是否随高度逐渐变透明
- overlap：控制重叠程度
- color：颜色
- hist&bin:是否将数据绘制成直方图


## 1.3 案例

```python
import seaborn as sns
import matplotlib.pyplot as plt
import joypy
import warnings

# 忽略警告
warnings.filterwarnings('ignore')
tips = sns.load_dataset('tips')
```

```python
sns.set_style('darkgrid')

plt.figure(figsize=(10,10))

joypy.joyplot(
    data=tips,
    column='total_bill',
    by='day'
)
plt.grid(True)
```

## 1.4 fade 是否随高度变透明

```python
sns.set_style('darkgrid')

plt.figure(figsize=(10,10))

joypy.joyplot(
    data=tips,
    column='total_bill',
    by='day',
    fade=True
)
plt.grid(True)
```

## 1.5 overlap控制重叠

默认=1，越小，间距越大

```python
sns.set_style('darkgrid')

plt.figure(figsize=(10,10))

joypy.joyplot(
    data=tips,
    column='total_bill',
    by='day',
    overlap=2,
)
plt.grid(True)
```

## 1.6 color设置颜色

颜色清单

![](../picture/B-kdeplot-1.png)

```python
sns.set_style('darkgrid')

plt.figure(figsize=(10,10))

joypy.joyplot(
    data=tips,
    column='total_bill',
    by='day',
    color='indigo',
    fade=True,
)
plt.grid(True)
```

## 1.7 hist&bins是否绘制为直方图,bins控制直方图的箱数

```python
sns.set_style('darkgrid')

plt.figure(figsize=(10,10))

joypy.joyplot(
    data=tips,
    column='total_bill',
    by='day',
    color='indigo',
    fade=True,
    overlap=0,
    hist=True,
    bins=10,
)
plt.grid(True)
```
