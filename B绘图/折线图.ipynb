```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = sns.load_dataset('flights')
data
```

```python
# 加载数据集

# 设置Seaborn风格和颜色调色板
sns.set_style('whitegrid')
# 设置图片大小,单位英寸
plt.figure(figsize=(10,10))

# 绘制折线图
sns.lineplot(
    data=data,
    x='year',
    y='passengers',
)
# 图标标题，坐标名
plt.title('year vs passengers')
plt.xlabel('year')
plt.ylabel('passengers')
plt.show()
```
