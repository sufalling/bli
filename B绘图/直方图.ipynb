# 1 直方图

Histogram用于展示数据的分布情况，通过将数据分组到连续的、不重叠的区间，以条形图的形式展现每个箱子中数据点的数量来表示数据的分布


使用场景

- 数据探索：数据分析的初步阶段，探索数据的分布特性
- 质量控制：制造业中分析产品尺寸等连续变量的质量分布
- 市场调研：市场分析中，用于展示消费者年龄、收入等连续变量的分布情况
- 金融风险控制：金融领域中用于分析投资回报率等连续变量的分布


## 1.1 sns.histplot

参数

- data:用于绘图的数据，DataFrame
- x:指定要绘制直方图的变量名
- y:如果提供，则绘制二维直方图
- bins:指定直方图的箱数，可以是整数（指定箱的数量），也可以是序列（指定具体的箱边界）
- kde:布尔值，决定是否在直方图上绘制概率密度估计曲线
- hue:分层变量，用于在直方图中创建多个子图，每个子图对应一个类别
- stat:控制直方图的y轴含义
- cumulative:控制是否对数据进行累加

## 1.2 直方图案例

### 1.2.1 基本直方图

```python
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
tips = sns.load_dataset('tips')
tips
```

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
sns.histplot(data=tips, x="tip")

plt.title('Distribution of Tips')
plt.xlabel('tips')
plt.ylabel('Frequency')

# 增加网格线
plt.grid(True)
#
plt.show()
```

### 1.2.2 bins指定直方图箱数



```python
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
sns.histplot(data=tips, x="tip", bins=30)

plt.title('Distribution of Tips')
plt.xlabel('tips')
plt.ylabel('Frequency')

# 增加网格线
plt.grid(True)
#
plt.show()
```

### 1.2.3 使用kde决定是否绘制概率密度估计曲线


```python
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
sns.histplot(data=tips, x="tip", bins=30,kde=True)

plt.title('Distribution of Tips')
plt.xlabel('tips')
plt.ylabel('Frequency')

# 增加网格线
plt.grid(True)
#
plt.show()
```

### 1.2.3 hue多子图直方图

两个直方图

```python
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
sns.histplot(data=tips, x="tip", bins=30,kde=True,hue="sex")

plt.title('Distribution of Tips')
plt.xlabel('tips')
plt.ylabel('Frequency')

# 增加网格线
plt.grid(True)
#
plt.show()
```

### 1.2.5 stat 控制直方图的y轴含义

- count:显示频数，默认
- density:显示概率密度，条形高度归一化，面积总和为1
- probability:用小数点表示的频率
- percent:用百分数表示的频率



```python
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
sns.histplot(data=tips, x="tip", bins=30,kde=True,stat="percent")


plt.title('Distribution of Tips')
plt.xlabel('tips')
plt.ylabel('Frequency')

# 增加网格线
plt.grid(True)
#
plt.show()
```

### 1.2.7 y变量二维直方图
