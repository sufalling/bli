# 1 气泡图

bubble plot是一种用于展示三个变量之间关系的图表类型，其中两个变量在水平和垂直轴上展示，第三个变量通过气泡的大小展示，这种图表非常适合展示数据集中的分布情况，尤其是当前需要强调某个特定数值的大小时

## 1.1 适用场景

- 市场研究：展示不同产品或服务的销售额、市场份额和价格
- 地理数据：展示不同地区的人口数量、经济规模等
- 科学研究：展示不同实验条件下的结果变量
- 金融分析：展示不同股票的市值、交易量和价格波动

## 1.2 sns.scatterplot

参数

- data:要绘制的数据集
- x:
- y:
- size:根据某一变量调整点的大小，size对于变量的值越大，点的大小越大
- sizes:控制点的大小范围，例如sizes=（10，100）可以让点的大小在20-200之间变化
- hue:用于分组，展示不同变量对应不同颜色
- palette:调色板
- alpha:气泡透明度


## 1.3 气泡图案例

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import title

tips = sns.load_dataset('tips')
tips
```

```python
sns.set_style('white')

plt.figure(figsize=(10,10))

sns.scatterplot(data=tips,x='total_bill',y='tip')

# plt.title()
# plt.legend()

plt.show()
```

## 1.3.1 size/sizes控制气泡大小

## 1.3.2 hue控制分组颜色


```python
sns.set_style('white')

plt.figure(figsize=(10, 10))

sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',size='size',sizes=(50,200))

# plt.title()
plt.legend(loc='upper right',title='sex',ncol=2)

plt.show()
```

## 1.3.4 palette颜色调色板

**Qualitative Palettes**：适合离散的类别数据，颜色彼此之间没有顺序或关联

- deep:默认，适合大部分情况
- muted:颜色较柔和，适合多类别对比
- pastel:颜色柔和
- dark:深沉的颜色，适合整数图表
- -colorblind：色盲友好，适合视觉障碍用户
- Set1，Set2，Set3，适合多个类别的颜色分配
- Paired：适合两两配对的颜色
- Accent：色调较鲜艳，适合强调特定类别


Sequential Paiettes：适合连续数据的颜色映射，从浅到深表示从大到小

## 1.3.5 alpha控制气泡透明度

```python
sns.set_style('white')

plt.figure(figsize=(10, 10))

sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',size='size',sizes=(50,200),alpha=0.8)

# plt.title()
plt.legend(loc='upper right',title='sex',ncol=2)

plt.show()
```
