# 1 气泡图

bubble plot是一种用于展示三个变量之间关系的图表类型，其中两个变量在水平和垂直轴上展示，第三个变量通过气泡的大小展示，这种图表非常适合展示数据集中的分布情况，尤其是当前需要强调某个特定数值的大小时

## 1.1 适用场景

- 市场研究：展示不同产品或服务的销售额、市场份额和价格
- 地理数据：展示不同地区的人口数量、经济规模等
- 科学研究：展示不同实验条件下的结果变量
- 金融分析：展示不同股票的市值、交易量和价格波动
