# 1 scatter plot散点图

一种展示两个数值变量关系的图表，二维平面上画点，展示变量之间的相关行、趋势、模式或异常值

适用场景：
- 探索变量关系：
- 趋势分析：增长，下降
- 模式识别：比如聚类和分组
- 异常值分析：异常值、离群点

## 1.1 sns.scatterplot()

参数：

- data:要绘制的数据集,DataFrame可以
- x:x轴数据
- y:y轴数据
- hue:用于分组的变量，比如为分类变量显示不同的颜色
- style:控制点形状，比如圆圈、方形等

## 1.2 案例
Iris鸢尾花数据集，包含三种鸢尾花的不同特征，包括花萼sepal长度，花萼宽带，petal花瓣长度和宽度、种类

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载Iris数据集
iris = sns.load_dataset('iris')
iris
```

### 1.2.1 基本散点图

- plt设置图片大小
- sns.scatterplot绘图
- plt.title,plt.xlabel,plt.ylabel设置图表标题和标签；plt.grid(True)设置网格线
- plt.show()显示渲染

plt.grid():
- True:是否画网格
- axis:设置显示哪个方向的网格线，可选值为'both'（默认），'x'或'y'
- linestyle:'--'
- color:颜色
- which:设置应用更改的网格线类型，可选值有'major'、'minor'和'both'，默认为'major'
- linewidth:宽度



```python
# 设置风格和颜色调色板,"white", "dark", "whitegrid", "darkgrid", "ticks
sns.set_style("white")
# 设置图片大小
plt.figure(figsize=(10,10))
# 绘制散点图，展示关系
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width")
# 设置图表标题和标签
plt.title('sepal_length vs sepal_width')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
# 网格线
plt.grid(True,axis='x')
# 渲染和显示
plt.show()
```

### 1.2.2 hue分组散点图

sns.scatterplot(....,hue='')

```python
plt.figure(figsize=(5,5))
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue='species')
plt.show()
```
