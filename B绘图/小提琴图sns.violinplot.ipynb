# 1 小提琴图

violin plot是箱型图和核密度图kde plot的结合，布局显示了数据的四分位数信息，还能直观地展示数据的分布情况，小提琴图适用于数据量较大且希望观察数据分布形态的场景

适用场景：

- 展示分布形状：需要展示数据分布形状，特别是是否呈现多峰分布
- 比较组分布：比较一个或多个分类变量不同级别的定量数据的分布
- 数据量巨大：数据量巨大，无法显示个别观察结果的情况


## 1.2 sns.violinplot参数

- data:数据源，DF或类似格式
- x:x轴变量
- y:y轴变量
- hue:分层变量，在小提琴图中创建多个子图
- split:布尔值，当为True，将hue参数定义的每个类别的小提琴图分开绘制，便于清晰比较不同组的分布
- palette:颜色方案
- alpha：透明度

## 1.3 小提琴图案例

tips

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tip = sns.load_dataset('tips')
tip
```

```python

sns.set_style('whitegrid')

plt.figure(figsize=(10,10))

sns.violinplot(data=tip,x='time',y='tip')

plt.title('violin plot----tip')
plt.xlabel('time')
plt.ylabel('tip')

plt.grid(True)

plt.show()
```

### 1.3.2 hue分类变量

#### 1.3.2.1 split=True,将小提琴一分为二
#### 1.3.2.1 split=False，不将小提琴一分为二

```python

sns.set_style('whitegrid')

plt.figure(figsize=(10,10))

sns.violinplot(data=tip,x='time',y='tip',hue='sex',split=True)

plt.title('violin plot----tip')
plt.xlabel('time')
plt.ylabel('tip')

plt.grid(True)

plt.show()
```

```python

sns.set_style('whitegrid')

plt.figure(figsize=(10,10))

sns.violinplot(data=tip,x='time',y='tip',hue='sex',split=False)

plt.title('violin plot----tip')
plt.xlabel('time')
plt.ylabel('tip')

plt.grid(True)

plt.show()
```

### 1.3.3 palette颜色参数

#### 1.3.3.1 常见颜色调色板palette介绍

colors=sns.color_palette("Set2")

**Qualitative Palettes**：适合离散的类别数据，颜色彼此之间没有顺序或关联

- deep:默认，适合大部分情况
- muted:颜色较柔和，适合多类别对比
- pastel:颜色柔和
- dark:深沉的颜色，适合整数图表
- colorblind：色盲友好，适合视觉障碍用户
- Set1，Set2，Set3，适合多个类别的颜色分配
- Paired：适合两两配对的颜色
- Accent：色调较鲜艳，适合强调特定类别


Sequential Paiettes：适合连续数据的颜色映射，从浅到深表示从大到小

### 1.3.4 水平小提琴图

x，y变量对换

```python

sns.set_style('whitegrid')

plt.figure(figsize=(10,10))

sns.violinplot(data=tip,y='time',x='tip',hue='sex',split=False)

plt.title('violin plot----tip')
plt.xlabel('time')
plt.ylabel('tip')

plt.grid(True)

plt.show()
```

### 1.3.5 小提琴图和箱线图结合


先sns.violinplot，再sns.boxplot最后一起设置标题标签和渲染

```python

sns.set_style('whitegrid')

plt.figure(figsize=(10,6))

sns.violinplot(data=tip,x='time',y='tip',split=False,alpha=0.5)
sns.boxplot(data=tip,x='time',y='tip',width=0.3)

plt.title('violin plot----tip')
plt.xlabel('time')
plt.ylabel('tip')

plt.grid(True)

plt.show()
```
