# 1 联合分布图

sns.joinplot绘制联合分布图的函数，显示两个变量之间的关系以及各自的分布情况，不仅展示两个变量之间的散点关系，在图的边缘还显示了每个变量的边际分布

使用场景
- 探索变量分布之间的关系：用来衡量两个连续变量之间的相关性、趋势和关系
- 分析分布，通过边际分布图展示每个变量的分布情况，帮助理解数据的偏态、分布特征和集中趋势
- 比较不同类别之间的关系，使用hue参数为不同类别的数据点上色，帮助比较不同类别在两个变量之间的关系差异

## 1.1 sns.jointplot

- x:
- Y:
- data:
- hue:按类别分布绘制
- height:控制联合分布图大小
- kind:指定绘图类型

## 1.2 数据集

![img](../picture/B-jointplot-1.png)

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset('penguins')
penguins
```

### 1.2.1 基本联合分布图


plt.suptitle设置图表总标题,加入y参数,往上偏移多少单位，不然总标题和图重叠

```python
sns.set_style('darkgrid')

sns.jointplot(
    x='flipper_length_mm',
    y='body_mass_g',
    data=penguins,
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)
# 普通标题会被上面的边际分布图盖住
# plt.title('jointplot',fontsize=20,fontweight='bold',y=1.02)

```

### 1.2.2 hue分组

### 1.2.3 height控制联合分布图大小(正方形)

```python
sns.set_style('darkgrid')

sns.jointplot(
    x='flipper_length_mm',
    y='body_mass_g',
    data=penguins,
    hue='species',
    height=10
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)
```

### 1.2.4 kind指定绘图类型

- scatter:散点，默认
- kde:绘制和密度估计图
- hist:绘制直方图
- reg:绘制带有回归线的散点图,不能带hue参数，会报错
- hex:绘制六边形热图,不能带hue参数，会报错

```python
sns.set_style('darkgrid')

sns.jointplot(
    x='flipper_length_mm',
    y='body_mass_g',
    data=penguins,
    hue='species',
    height=10,
    kind='kde',
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)
```

```python
sns.set_style('darkgrid')

sns.jointplot(
    x='flipper_length_mm',
    y='body_mass_g',
    data=penguins,
    hue='species',
    height=10,
    kind='hist',
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)
```

```python
sns.set_style('darkgrid')

sns.jointplot(
    x='flipper_length_mm',
    y='body_mass_g',
    data=penguins,
    height=10,
    kind='reg',
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)
```

```python
sns.set_style('darkgrid')

sns.jointplot(
    x='flipper_length_mm',
    y='body_mass_g',
    data=penguins,
    height=10,
    kind='hex',
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)
```
