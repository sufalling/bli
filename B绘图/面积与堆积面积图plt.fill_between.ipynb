# 1 面积图

Area plot 是一种通过填充曲线与轴之间的区域来可视化数据的图形，是折线图的变种，将折线图下方的区域进行填充，通常用于展示某个时间段不同类别数据的累积量变化

使用场景

- 展示累计值变化：适用于展示随时间或其他变量变化的累计值，如总收入总用户数
- 显示部分对整体的贡献：堆叠多个数据系列，直观展示各部分对整体的相对贡献，
- 比较多个数据系列的变化趋势
- 表现数据的上升或下降趋势



## 1.1 plt.fill_between

参数

- x：x轴变量
- y1:y轴
- data:DF
- color:线条颜色
- alpha:填充的透明度
- label:图例标签
- hatch:填充区域的纹理样式



## 1.2 面积图案例

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 生成模拟数据，模拟GDP
data = {
    'year':list(range(2000,2020)),
    'USA':list(np.random.uniform(0,20,20)),
    'china':list(np.random.uniform(0,20,20)),
    'Germany':list(np.random.uniform(0,20,20)),
}
gdp = pd.DataFrame(data)
gdp
```

### 1.2.1 基本面积图

只需要x，y1，data三个参数

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
plt.fill_between(x='year',y1='china',data=gdp)

plt.title('area plot GDP:billion)')
plt.xlabel('year')
plt.ylabel('china')

#
plt.show()
```

### 1.2.2 多个面积图

类似，使用多个fill_between

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
plt.fill_between(x='year',y1='china',data=gdp)
plt.fill_between(x='year',y1='Germany',data=gdp)
plt.fill_between(x='year',y1='USA',data=gdp)

plt.title('area plot GDP:billion)')
plt.xlabel('year')
plt.ylabel('china')

#
plt.show()
```

### 1.2.3 color 颜色设置

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
plt.fill_between(x='year',y1='china',data=gdp,color='red')
plt.fill_between(x='year',y1='Germany',data=gdp,color='blue')
plt.fill_between(x='year',y1='USA',data=gdp,color='green')

plt.title('area plot GDP:billion)')
plt.xlabel('year')
plt.ylabel('china')

#
plt.show()
```

### 1.2.4 alpha透明度

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
plt.fill_between(x='year',y1='china',data=gdp,color='red',alpha=0.5)
plt.fill_between(x='year',y1='Germany',data=gdp,color='blue',alpha=0.5)
plt.fill_between(x='year',y1='USA',data=gdp,color='green',alpha=0.5)

plt.title('area plot GDP:billion)')
plt.xlabel('year')
plt.ylabel('china')

#
plt.show()
```

### 1.2.5 label设置标签

最后加plt.legend(title)

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
plt.fill_between(x='year',y1='china',data=gdp,color='red',alpha=0.5,label='china')
plt.fill_between(x='year',y1='Germany',data=gdp,color='blue',alpha=0.5,label='Germany')
plt.fill_between(x='year',y1='USA',data=gdp,color='green',alpha=0.5,label='USA')

plt.title('area plot GDP:billion)')
plt.xlabel('year')
plt.ylabel('china')
plt.legend(title='Country')
#
plt.show()
```

### 1.2.6 hatch填充图案

- /:斜线填充
- \\\:反斜线填充
- |:竖直线
- -:水平线
- +:十字
- .:点
- x：斜交叉线
- *；星形
- `倒引号

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
plt.fill_between(x='year',y1='china',data=gdp,color='red',alpha=0.5,label='china',hatch='x')
plt.fill_between(x='year',y1='Germany',data=gdp,color='blue',alpha=0.5,label='Germany',hatch='+')
plt.fill_between(x='year',y1='USA',data=gdp,color='green',alpha=0.5,label='USA',hatch='-')

plt.title('area plot GDP:billion)')
plt.xlabel('year')
plt.ylabel('china')
plt.legend(title='Country')
#
plt.show()
```
