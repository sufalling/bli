# 1 密度图sns.kdeplot

Density plot 是一种用于展示连续变量概率分布的图形，通过平滑的曲线来近似表示数据的概率密度函数，从而揭示数据的分布特征，如单峰、双峰、偏斜等，常用于数据探索阶段，直观地理解数据的形状和特征

使用场景

- 单变量分析：分析单个连续变量的分布情况
- 双变量分析：比较两个连续变量的分布差异
- 模式识别：识别数据中的异常值、聚焦或间隙

## 1.1 sns.kdeplot

参数：

- data:DataFrame，数据来源
- x:x轴数据名称
- y:y轴数据名称
- fill:是否填充曲线下方区域
- color:字符串或颜色代码，指定曲线颜色
- alpha:控制曲线透明度
- hue:根据某个分类变量绘制不同颜色的密度曲线


## 1.2 密度图案例

tips小费数据集

### 1.2.1 密度图基本案例

```python
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
tips = sns.load_dataset('tips')
tips
```

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 基本密度图，只需要data和x
sns.kdeplot(
    data=tips,
    x='tip'
)

# 标题
plt.title('tip density',fontsize=20,fontweight='bold')
# 网格线
plt.grid(True)
#
plt.show()
```

### 1.2.2 fill填充曲线下方区域


```python
sns.kdeplot(
    data=tips,
    x='total_bill',
    fill=True
)

plt.show()
```

### 1.2.3 color 颜色参数

![color](../picture/B-kdeplot-1.png)

```python
sns.kdeplot(
    data=tips,
    x='total_bill',
    fill=True,
    color='cyan'
)

plt.show()
```

### 1.2.4 alpha控制曲线透明度，0-1.0，从透明到不透明

```python
sns.kdeplot(
    data=tips,
    x='total_bill',
    fill=True,
    alpha=0.3
)

plt.show()
```

### 1.2.5 hue分组绘制多子图密度图

```python
sns.kdeplot(
    data=tips,
    x='total_bill',
    fill=True,
    hue='sex'
)

plt.show()
```

```python
sns.kdeplot(
    data=tips,
    x='total_bill',
    fill=True,
    hue='smoker'
)

plt.show()
```

### 1.2.6 y-双变量密度图

```python
sns.kdeplot(
    data=tips,
    x='total_bill',
    y='tip',
    fill=True,
    alpha=0.6
)

plt.show()
```
