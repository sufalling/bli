# 1 双坐标轴图

同一个图中使用两个不同的y轴，两个y轴可以有不同的尺度和单位，用于展示两种具有不同量纲或范围的数据，且两种数据共享一个x轴

## 1.1 plt.twinx():创建第二个坐标轴

## 1.2 案例

1967-2015年美国经济关键指标数据

![img](../picture/B-twinx-1.png)

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/selva86/datasets/raw/master/economics.csv')

# 转换为时间
df['date'] = pd.to_datetime(df['date'])
df
```

## 1.2.1 双折线图

## 1.2.2 散点+折线

## 1.2.3 散点+散点

```python
sns.set_style("white")
plt.figure(figsize=(10,10))

sns.lineplot(data=df, x='date', y='pce',color='blue',label='pce')

# 创建第二个坐标轴

plt.twinx()

# 绘制第二个坐标轴
sns.lineplot(data=df, x='date', y='psavert',color='red',label='psavert')

plt.legend()
```

```python
sns.set_style("white")
plt.figure(figsize=(10,10))

sns.scatterplot(data=df, x='date', y='pce',color='blue',label='pce')

# 创建第二个坐标轴

plt.twinx()

# 绘制第二个坐标轴
sns.lineplot(data=df, x='date', y='psavert',color='red',label='psavert')

plt.legend()
plt.show()
```
