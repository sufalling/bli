# 1 点图

strip plot 将每个数据点沿着一个轴方向排列，与散点图相比，适合对离散数据的可视化展示


## 1.1 使用场景


- 离散数据展示：可以展示每个类别下的数值分布情况
- 查看离群点：每个数据点单独展示，可帮助识别异常值、离群点
- 适合小型数据集：精确展示每个数据点


## 1.2 sns.stripplot


参数

- data:数据集。DF
- x:
- y:
- hue:用于分组的变量
- jitter:是否在x轴上加入轻微的随机偏移，使数据点不会完全重叠，设置为True表示加入偏移
- dodge:如果hue参数启用，dodge=True会将数据点根据hue分类分开防止重叠，默认false
- palette:调色板
- size:数据点大小，默认为5

## 1.2 案例

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')


sns.set_style("white")
plt.figure(figsize=(10,10))

sns.stripplot(data=tips, x='day', y='total_bill')
```

## 2.1 hue分组变量

```python
tips = sns.load_dataset('tips')


sns.set_style("white")
plt.figure(figsize=(10,10))

sns.stripplot(data=tips, x='day', y='total_bill',hue='sex')
plt.savefig('../test/fig1')
```

## 2.2 jitter加入随机偏移，默认True

## 2.3 dodge是否根据hue将数据点分开



```python
tips = sns.load_dataset('tips')


sns.set_style("white")
plt.figure(figsize=(10,10))

sns.stripplot(data=tips, x='day', y='total_bill',hue='sex',dodge=True)
```

## 2.4 常见颜色调色板palette介绍

colors=sns.color_palette("Set2")

**Qualitative Palettes**：适合离散的类别数据，颜色彼此之间没有顺序或关联

- deep:默认，适合大部分情况
- muted:颜色较柔和，适合多类别对比
- pastel:颜色柔和
- dark:深沉的颜色，适合整数图表
- -colorblind：色盲友好，适合视觉障碍用户
- Set1，Set2，Set3，适合多个类别的颜色分配
- Paired：适合两两配对的颜色
- Accent：色调较鲜艳，适合强调特定类别


Sequential Paiettes：适合连续数据的颜色映射，从浅到深表示从大到小

## 2.5 size设置点大小

```python
tips = sns.load_dataset('tips')


sns.set_style("white")
plt.figure(figsize=(10,10))

sns.stripplot(data=tips, x='day', y='total_bill',hue='sex',dodge=True,size=30)
```
