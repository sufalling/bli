# 1 pairplot

sns.pairplot绘制每对变量之间的联合分布图，在对角线位置绘制每个变量的单独分布图，从而展示多个变量之间的关系

适用场景

- 探索多个变量之间的关系：快速了解多个连续变量之间的相互关系
- 揭示数据相关性、分布和结构：
- 查看类别之间的差异：hue参数分组绘制
- 检测异常值：潜在的异常或离群点，尤其是包含多个变量时

## 1.2 sns,pairplot

- data:数据
- hue:根据类别着色
- markers:散点样式
- palette:调色板
- plot_kws:散点样式
- kind:非对角线图类型
- diag_kind:对角线图形类型


## 1.2 案例



```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset('penguins')
penguins
```

### 1.2.1 基本配对图

```python
sns.set_style('darkgrid')

sns.pairplot(
    data=penguins,
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)

plt.show()
```

### 1.2.3 markers散点样式


使用字典为不同类别设置不同样式

```python
sns.set_style('darkgrid')

sns.pairplot(
    data=penguins,
    hue='species',
    markers={'Adelie':'o','Chinstrap':'X','Gentoo':'D'},
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)

plt.show()
```

### 1.2.4 常见颜色调色板palette介绍

colors=sns.color_palette("Set2")

**Qualitative Palettes**：适合离散的类别数据，颜色彼此之间没有顺序或关联

- deep:默认，适合大部分情况
- muted:颜色较柔和，适合多类别对比
- pastel:颜色柔和
- dark:深沉的颜色，适合整数图表
- -colorblind：色盲友好，适合视觉障碍用户
- Set1，Set2，Set3，适合多个类别的颜色分配
- Paired：适合两两配对的颜色
- Accent：色调较鲜艳，适合强调特定类别


Sequential Paiettes：适合连续数据的颜色映射，从浅到深表示从大到小

![Sequential](../picture/B-pieplot-1.png)

![Sequential](../picture/B-pieplot-2.png)

```python
sns.set_style('darkgrid')

sns.pairplot(
    data=penguins,
    hue='species',
    markers={'Adelie':'o','Chinstrap':'X','Gentoo':'D'},
    palette='Set1'
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)

plt.show()
```

### 1.2.5 plot_kws设置散点样式

- s:大小
- alpha:透明度

```python
sns.set_style('darkgrid')

sns.pairplot(
    data=penguins,
    hue='species',
    markers={'Adelie':'o','Chinstrap':'X','Gentoo':'D'},
    plot_kws={'alpha':0.8,'s':14},
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)

plt.show()
```

### 1.2.6 kind指定联合分布图的类型

- scatter:散点
- kde:核密度图
- hist:直方图

### 1.2.7 diag_kind 指定对角线图形的类型

- kde:核密度图
- hist:直方图

```python
sns.set_style('darkgrid')

sns.pairplot(
    data=penguins,
    hue='species',
    markers={'Adelie':'o','Chinstrap':'X','Gentoo':'D'},
    # plot_kws={'alpha':0.8,'s':14},
    kind='hist',
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)

plt.show()
```

```python
sns.set_style('darkgrid')

sns.pairplot(
    data=penguins,
    hue='species',
    markers={'Adelie':'o','Chinstrap':'X','Gentoo':'D'},
    plot_kws={'alpha':0.8,'s':14},
    diag_kind='hist'
)

# 设置图表总标题
plt.suptitle('jointplot',fontsize=20,fontweight='bold',y=1.02)

plt.show()
```
