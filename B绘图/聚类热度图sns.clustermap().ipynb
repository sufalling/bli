# 1 聚类热度图

聚类分析+热度图，对聚类进行可视化

构成
- 热度图：色块的颜色表示数据大小
- 聚类：相似的行或列聚合在一起

## 1.1 sns.clustermap()


- data:数据
- cmap:选择颜色映射方案
- method:选择聚类算法
- metric:选择计算距离的方法
- annot:控制是否在每个单元格显示数值
- fmt:选择数值显示格式

## 1.2 案例

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# 创建示例数据

df = pd.DataFrame(
    data=np.random.rand(10,10),
    columns=[f'stock_{i}' for i in range(10)],
    index=[f'day_{i}' for i in range(10)]
)
df
```

## 1.3 基本聚类热度图


```python
sns.set_style("white")
plt.figure(figsize=(10,10))

sns.clustermap(data=df)
```

## 1.3.2 cmap颜色映射

![img](../picture/B-clustermap-1.png)
![img](../picture/B-clustermap-2.png)

```python
sns.set_style("white")
plt.figure(figsize=(10,10))

sns.clustermap(data=df,cmap="YlGnBu")
```

### 1.3.3 method聚类方法介绍

![img](../picture/B-clustermap-3.png)

```python
sns.set_style("white")
plt.figure(figsize=(10,10))

sns.clustermap(data=df,cmap="YlGnBu",method="ward")
```

### 1.3.5 metric计算距离方法,annot显示数值

![img](../picture/B-clustermap-4.png)
![img](../picture/B-clustermap-5.png)

```python
sns.set_style("white")
plt.figure(figsize=(10,10))

sns.clustermap(data=df,cmap="YlGnBu",method="ward",metric="euclidean",annot=True)
```

### 1.3.7 fmt控制格式

- .nf:保留小数
- .ne:保留几位科学计数法

```python
sns.set_style("white")
plt.figure(figsize=(10,10))

sns.clustermap(data=df,cmap="YlGnBu",method="ward",annot=True,fmt=".3f")
```
