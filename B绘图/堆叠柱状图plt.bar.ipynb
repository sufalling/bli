# 1 堆叠柱状图

stacked bar chart是一种数据可视化图标，将不同类别的数据值堆叠在一起，展示不同类别直接的比较和各自的总量。每个柱子代表一个总量，每个柱子的不同颜色代表构成这个总量的各个部分的值，这种图表适合展示各个部分对整体的贡献以及随时间或不同条件的变化

使用场景

- 展示总量与部分量：展示一个总量由几个部分组成，各部分对总量的贡献
- 比较不同类别，在不同条件或不同时间段的表现
- 展示比例关系


## 1.1 plt.bar

参数

- x:控制柱子的x轴位置，类别
- height:控制柱子高度，y轴数量
- bottom:控制柱子的底部起始位置
- color:设置柱子颜色
- hatch:设置填充图案
- label：设置图例标签


## 1.2 案例



```python
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
# 创建日期范围
dates = pd.date_range('20230101',periods=7)

# 创建数据集

date = {
    'Date': dates,
    'Products_A':np.random.randint(1,10,len(dates)),
    'Products_B':np.random.randint(1,10,len(dates)),
    'Products_C':np.random.randint(1,10,len(dates)),
    'Region':np.random.choice(['a','b','c','d'],len(dates)),
}

df = pd.DataFrame(date)
df
```

### 1.2.1 基本堆叠柱状图绘制



```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 初始化底部数组,指定每个大柱子的起始位置,第一个产品是紧靠x轴所以全是0
bottom = np.zeros(len(df['Date']))
# 绘制产品A的柱状图
plt.bar(x=df['Date'], height=df['Products_A'], bottom=bottom)
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_A']
# 下面类似
# 绘制产品B的柱状图
plt.bar(x=df['Date'], height=df['Products_B'], bottom=bottom)
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_B']
# 绘制产品C的柱状图
plt.bar(x=df['Date'], height=df['Products_C'], bottom=bottom)
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_C']

# 设置标题和轴名


# 调整x轴标签的旋转角度
plt.xticks(rotation=45)
# 绘制
plt.show()
```

### 1.2.2 color设置柱子颜色





```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 初始化底部数组,指定每个大柱子的起始位置,第一个产品是紧靠x轴所以全是0
bottom = np.zeros(len(df['Date']))
# 绘制产品A的柱状图
plt.bar(x=df['Date'], height=df['Products_A'], bottom=bottom,color='red')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_A']
# 下面类似
# 绘制产品B的柱状图
plt.bar(x=df['Date'], height=df['Products_B'], bottom=bottom,color='blue')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_B']
# 绘制产品C的柱状图
plt.bar(x=df['Date'], height=df['Products_C'], bottom=bottom,color='yellow')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_C']

# 设置标题和轴名


# 调整x轴标签的旋转角度
plt.xticks(rotation=45)
# 绘制
plt.show()
```


### 1.2.3 hatch设置填充图案




```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 初始化底部数组,指定每个大柱子的起始位置,第一个产品是紧靠x轴所以全是0
bottom = np.zeros(len(df['Date']))
# 绘制产品A的柱状图
plt.bar(x=df['Date'], height=df['Products_A'], bottom=bottom,color='red',hatch='/')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_A']
# 下面类似
# 绘制产品B的柱状图
plt.bar(x=df['Date'], height=df['Products_B'], bottom=bottom,color='blue',hatch='-')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_B']
# 绘制产品C的柱状图
plt.bar(x=df['Date'], height=df['Products_C'], bottom=bottom,color='yellow',hatch='.')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_C']

# 设置标题和轴名


# 调整x轴标签的旋转角度
plt.xticks(rotation=45)
# 绘制
plt.show()
```


### 1.2.4 label设置图例标签


plt.bar设置完label标签，最后添加plt.legend(title='title')

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 初始化底部数组,指定每个大柱子的起始位置,第一个产品是紧靠x轴所以全是0
bottom = np.zeros(len(df['Date']))
# 绘制产品A的柱状图
plt.bar(x=df['Date'], height=df['Products_A'], bottom=bottom,color='red',hatch='/',label='a')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_A']
# 下面类似
# 绘制产品B的柱状图
plt.bar(x=df['Date'], height=df['Products_B'], bottom=bottom,color='blue',hatch='-',label='b')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_B']
# 绘制产品C的柱状图
plt.bar(x=df['Date'], height=df['Products_C'], bottom=bottom,color='yellow',hatch='.',label='c')
# 更新bottom因为下一个产品的底部是A产品的顶部
bottom += df['Products_C']

# 设置标题和轴名


# 设置图例
plt.legend(title='Products')
# 调整x轴标签的旋转角度
plt.xticks(rotation=45)
# 绘制
plt.show()
```
