# 1 热力图

heatmap是通过颜色强度来表示数据的二维图形，常用于展示数值之间的关系或数据的分布趋势，热力图能够帮助快速识别出数据中的模式趋势和异常

使用场景

- 数据分布展示，展示二维数据的分布，如人口密度分析、活跃度分析
- 相关性分析，展示变量之间的相关系数矩阵，通过颜色的强度来表示相关性的强弱
- 地理信息展示，展示地理位置的数据密度，如人群分布，温度变化等


## 1.1 sns.heatmap

参数

- data:数据源，DF或类似格式
- annot:布尔值。控制是否在每个单元格显示数值
- fmt:字符串，指定annot参数中数值的格式化字符串
- cmap:字符串或matplotlib颜色映射对象，指定热力图的颜色映射方案
- cbar:布尔值，设置是否显示颜色条
- center:设置颜色条的中心值

## 1.2 热力图案例

titanic数据集，包含891条乘客记录，记录了个人信息和titanic号上的生存情况

![img](../picture/B-heatmap-1.png)

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
# 筛选数值型变量
numeric_titanic = titanic.select_dtypes(include=['number'])
corr = numeric_titanic.corr()
corr
```

### 1.2.1 基本热力图

只使用数据源得到的相关系数矩阵即可

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
sns.heatmap(data=corr,)

plt.title('heatmap')

#
plt.show()
```

### 1.3.2 annot设置是否在每个单元格显示数值

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
sns.heatmap(data=corr,annot=True)

plt.title('heatmap')

#
plt.show()
```

### 1.3.3 fmt设置数值显示格式

语法

- 保留小数位数：
    - ‘.2f':保留两位小数，f表示浮点数
- 科学计数法
    - ’.1e'：使用科学计数法，保留一位有效数字
    - ’.3e'：使用科学计数法，保留三位有效数字

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
sns.heatmap(data=corr,annot=True,fmt='.4e')

plt.title('heatmap')

#
plt.show()
```

### 1.2.4 cmap设置调色板

6种颜色组合

![](../picture/B-heatmap-2.png)

```python
# 展示调色板
sns.palplot(sns.color_palette('viridis'))
```

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
sns.heatmap(data=corr,annot=True,fmt='.4e',cmap='YlGnBu')

plt.title('heatmap')

#
plt.show()
```

### 1.2.5 cbar设置是否显示颜色条

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
# sns.heatmap(data=corr,annot=True,fmt='.4e',cmap='YlGnBu',cbar=False)
sns.heatmap(data=corr,annot=True,fmt='.4e',cmap='YlGnBu',cbar=True)
plt.title('heatmap')
plt.show()
```

### 1.2.6 center设置颜色条的中心值

例如，设置center=0，0会处于颜色条的中间位置，正负数值会根据0对称分布

```python
# 设置风格
sns.set_style("white")

# 设置图片大小
plt.figure(figsize=(12, 7))

# 绘制直方图，只需要data和x即可
# sns.heatmap(data=corr,annot=True,fmt='.4f')
sns.heatmap(data=corr,annot=True,fmt='.4e',cmap='YlGnBu',cbar=True,center=-1)

plt.title('heatmap')

#
plt.show()
```
